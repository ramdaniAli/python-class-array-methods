#### Code the class 'Aquarium' that outputs the result as shown. 

```python
class Aquarium():
 
    def __init__(self):
        print("An Aquarium has opened!")
        self.aqua = []
    
    
    def enters(self , fish):
        if(fish in self.aqua):
            print('We already have '+ fish)
            return 
        else : 
            print(fish + ' is a new member of our Aquarium!')
            return self.aqua.append(fish)
    
    def leaves(self , fish):
        if(fish in self.aqua):
            return self.aqua.remove(fish)
        else : 
            print('No such sea creature in the Aquarium!') 
            return

    def show_all(self):
        return self.aqua
    
```
